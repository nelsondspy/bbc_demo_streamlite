```python
import pandas as pd 
from os.path import join as pjoin 
import os 



data = './data'
dataset_name = 'bbc_full_dataset.pkl'


def load_catalog_data():
    files = ['arts','cbbc','comedy','documentaries','entertainment','films','from-the-archives',
             'history','lifestyle','music','sports','signed','science-and-nature']
    frames = []
    for f in files:
        df  =  pd.read_pickle(pjoin(data, f + '.pkl'))    
        frames.append(df)
        print(f,":",  len(df))
    return frames 
    
catalog_dataset = None 

if os.path.exists(dataset_name):
    content_dataset = pd.read_pickle(dataset_name)
else:    
    catalog_dataset = pd.concat(load_catalog_data())
    catalog_dataset['id'] = catalog_dataset.index
    catalog_dataset.to_pickle(pjoin(data, dataset_name))

catalog_dataset.head(5)
#catalog_dataset['show'] 
#df_news = pd.read_pickle(pjoin(data,  'news.pkl'))
#df_news = catalog_dataset[catalog_dataset['show'].str.contains("Ozark")]
#df_news.sample(10).head()


```
